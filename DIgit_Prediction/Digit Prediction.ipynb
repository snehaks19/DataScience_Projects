<h2 style="color:blue">Logistic Regression: Multiclass Classification</h2>

show how to use sklearn logisticregression class to solve multiclass classification problem to predict hand written digit

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
dir(digits)      # list the names of all the attributes and methods available for an object.
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])  #matshow function is used to display a 2D array as an image
    plt.figure(figsize=(2,2))
```

```python
 digits.target[0:5]  #we can use data and target to train our model
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(digits.data,digits.target,test_size=0.2)        #for not overfit
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='lbfgs', max_iter=5000)
```

```python
model.fit(X_train,y_train)     #   xtrain has handwritten inout ,ytrain has output
```

```python
model.score(X_test,y_test)
```

```python
plt.matshow(digits.images[67])
```

```python
 digits.target[67]
```

```python
model.predict([digits.data[67]])
```

```python
model.predict(digits.data[0:5])
```

```python
y_predicted=model.predict(X_test)
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,y_predicted)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
